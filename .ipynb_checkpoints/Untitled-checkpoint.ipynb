Removing the loan_status of 'current' type as the analysis is identify the risks of defaulting the loans

Dropping columns which has high percentage of missing values and large unique value of string types as we cannot get a insights from these columns

```python

data_state_addr = df.reset_index().pivot_table(index='addr_state',columns='loan_status', values='loan_amnt',aggfunc='count')
data_state_addr['charged_off_per'] = data_state_addr['Charged Off'].astype(float)/(data_state_addr['Charged Off'].astype(float) + data_state_addr['Fully Paid'].astype(float) )
data_state_addr['Fully_Paid_per'] = 1 - data_state_addr['charged_off_per']
data_state_addr.plot.bar(y=['Fully_Paid_per','charged_off_per'],figsize=(20,16))
```

```python
# Categorising the continuse variables into a buckets for bivariate analysis

# Categorising the loan amount column
df['loan_amnt_category'] = pd.cut(df['loan_amnt'], [0, 5000, 10000, 20000, 25000, 35000], labels=['0-5000', '5000-10000', '10000-20000', '20000-25000', '25000 +'])

# Categorising the annual income  column
df['annual_inc_category'] = pd.cut(df['annual_inc'], [0,20000, 40000, 60000, 80000,1000000], labels=['0-20000', '20000-40000', '40000-60000', '60000-80000', '80000 +'])

# Categorising the interest rate  column
df['int_rate_category'] = pd.cut(df['int_rate'], [0, 10, 12.5, 16, 20], labels=['0-10', '10-13', '12.5-16', '16 +'])

# Categorising the interest rate  column
df['dti_category'] = pd.cut(df['dti'], [0, 5, 10, 15, 20, 25], labels=['0-5', '05-10', '10-15', '15-20', '25+'])
```

```python
data_annual_income_loan = df.groupby(['annual_inc_category', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()


data_annual_income_loan['charged_off_per'] = data_annual_income_loan['Charged Off'].astype(float)/(data_annual_income_loan['Charged Off'].astype(float) + data_annual_income_loan['Fully Paid'].astype(float) )

data_annual_income_loan['full_paid_per'] = 1-data_annual_income_loan['charged_off_per'] 
data_annual_income_loan.plot.bar(x='annual_inc_category', y=['full_paid_per','charged_off_per'],figsize=(20,16))
```

Borrowers with low income are more likley to default

```python
loan_amnt_category_loan_status = df.groupby(['loan_amnt_category', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()


loan_amnt_category_loan_status['charged_off_per'] = loan_amnt_category_loan_status['Charged Off'].astype(float)/(loan_amnt_category_loan_status['Charged Off'].astype(float) + loan_amnt_category_loan_status['Fully Paid'].astype(float) )
loan_amnt_category_loan_status['full_paid_per'] = 1-loan_amnt_category_loan_status['charged_off_per'] 
loan_amnt_category_loan_status.plot.bar(x='loan_amnt_category', y=['full_paid_per','charged_off_per'],figsize=(20,16))
```
